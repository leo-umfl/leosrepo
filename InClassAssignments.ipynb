```python
sales = float(input("Sales Amount: "))
tax = float(input("Tax Rate (%): "))
cogs = sales*.4
grossprofit = sales-cogs
taxes = grossprofit*(tax/100)
netincome = grossprofit-taxes

print(f"Sales: {sales:.2f}")
print(f"COGS: {cogs:.2f}")
print(f"Gross Profit: {grossprofit:.2f}")
print(f"Taxes: {taxes:.2f}")
print(f"Net Income: {netincome:.2f}")
```

```python
number1 = float(input("First Number: "))
operator = (input("Operator: "))
number2 = float(input("Second Number: "))

if operator == "+":
  res = number1+number2
elif operator == "-":
  res = number1-number2
elif operator == "*":
  res = number1*number2
elif operator == "/":
  res = number1/number2
elif operator == "//":
  res = number1//number2
elif operator == "**":
  res = number1**number2
elif operator == "%":
  res= number1%number2
else:
  print("invalid operation")

print(f"Result: {res}")
```

```python
import random

play = input("Do you want to play a game (y/n)? ")

while play == 'y':
  player = input("Choose b/w rock, paper, scissors: ").lower()
  computer = random.choice(['rock', 'paper', 'scissors'])

  if player == computer:
    print("Draw")
  elif player == "rock" and computer == "scissors":
    print("Player wins!")
  elif player == "scissors" and computer == "paper":
    print("Player wins!")
  elif player == "paper" and computer == "rock":
    print("Player wins!")
  else:
    print ("Computer wins")

  play = input("Do you want to continue (y/n)? ")
```

```python
def getValues(n):
  assert isinstance (n, int) and n>0, "n must be a positive integer"
  values = []
  for i in range(n):
    try:
      val = float(input("Enter value: "))
      values.append(val)
    except ValueError:
      print("Warning: Invalid Input. Please enter number")
  return values

def calcAverage(alist):
  assert isinstance(alist, list), "Input must be a list"
  if len(alist)<2:
    raise Exception("List must contain at least two values")
  included = alist[2:]
  return sum(included)/len(included)

try:
  n=int(input("How many values? "))
  values=getValues(n)
  print("Average:",calcAverage(values))
except ValueError:
  print("Warning: Numbers must be integers")
except Exception as e:
  print("Error:",e)
```

```python
class Greeting:
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name

    def display(self, formal):
        if formal:
            print(f"{self.last_name}, {self.first_name}")
        else:
            print(self.first_name)

    def __str__(self):
        return "Class Greeting"


p = Greeting("Leo", "Spanier")
print(p)
p.display(formal=False)
p.display(formal=True)
```

```python
class Student:
  def __init__(self, name, number):
    self.__name=name
    self.number=number
    self.__courses=[]

  @property
  def name(self):
    return self.__name

  @name.setter
  def name(self,newName):
    self.name=name

  def get_courses(self):
    return self.__courses

  def __len__(self):
    return len(self.__courses)

  def enroll(self, newCourse):
    if newCourse not in self.__courses:
      self.__courses.append(newCourse)
    else:
      print(f"Student {self.__name} has already enrolled in {newCourse}.")
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance(self, other):
        dx=(self.x-other.x)**2
        dy=(self.y-other.y)**2
        return (dx+dy)**0.5

class Rectangle:
    def __init__(self):
        self.p1 = Point(0,0)
        self.p2 = Point(3,0)
        self.p3 = Point(0,3)
        self.p4 = Point(3,3)

        self.width = self.p1.distance(self.p2)
        self.height = self.p1.distance(self.p3)

    def area(self):
        return self.width * self.height
```

```python
z=Rectangle()
z.area()
```

```python
class Employee:
  def __init__(self,name,wage,hours):
    self.name=name
    self.__hourly_wage=wage
    self.hours=hours

  def get_wage(self):
    return self.__hourly_wage

  def set_wage(self,newWage):
    if newWage<=0:
      print("Negative wage values not allowed.")
    else:
     self.__hourly_wage=newWage

  def calculate_pay(self):
    return self.__hourly_wage*self.hours

  def __str__(self):
    return f"Employee: {self.name}\nHourly wages: {self.__hourly_wage}$/hour"

class Manager(Employee):
  def __init__(self,name,wage,hours,bonus):
    super().__init__(name,wage,hours)
    self.bonus=bonus

  def calculate_pay(self):
    return (1+self.bonus)*self.get_wage()*self.hours

employee=Employee("Bob",50,10)
manager=Manager("Alice",100,15,.2)
print(employee)
print(manager)
print(f"Employee {employee.name} weekly salary is: {employee.calculate_pay()}$")
print(f"Manager {manager.name} weekly salary is: {manager.calculate_pay()}$ that includes a {manager.bonus *100}% bonus")
```
