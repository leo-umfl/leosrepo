1st ICA

```python
sales = float(input("Sales Amount: "))
tax = float(input("Tax Rate (%): "))
cogs = sales*.4
grossprofit = sales-cogs
taxes = grossprofit*(tax/100)
netincome = grossprofit-taxes

print(f"Sales: {sales:.2f}")
print(f"COGS: {cogs:.2f}")
print(f"Gross Profit: {grossprofit:.2f}")
print(f"Taxes: {taxes:.2f}")
print(f"Net Income: {netincome:.2f}")
```

2nd ICA

```python
number1 = float(input("First Number: "))
operator = (input("Operator: "))
number2 = float(input("Second Number: "))

if operator == "+":
  res = number1+number2
elif operator == "-":
  res = number1-number2
elif operator == "*":
  res = number1*number2
elif operator == "/":
  res = number1/number2
elif operator == "//":
  res = number1//number2
elif operator == "**":
  res = number1**number2
elif operator == "%":
  res= number1%number2
else:
  print("invalid operation")

print(f"Result: {res}")
```

3rd ICA

```python
print("Hour Fee")
for hour in range(1,9):
  fee=5+2.5*hour
  if fee < 10.0:
    fee=10.0
  elif fee>20.0:
    fee=20.0
  print(f"{hour}    {fee}")
```

4th ICA

```python
ticketPrice = 10
defaultAudience = 20

print(f"{"Ad$\tProfit"}")
for adMoney in range(0, 201, 25):
  revenue = (defaultAudience + 2 * adMoney ** 0.5)*ticketPrice
  costs = 200 + adMoney
  profit = revenue - costs
  print(f"{float(adMoney)}\t{float(profit)}")
```

5th ICA

```python
print("Hours\tFee")

for hours in range(1,9):
  fee=5+2.5*hours
  if fee < 10:
    fee=10
  if fee > 20:
    fee=20
  print(hours,"\t$", fee)
```

6th ICA

```python
import random

play = input("Do you want to play a game (y/n)? ")

while play == 'y':
  player = input("Choose b/w rock, paper, scissors: ").lower()
  computer = random.choice(['rock', 'paper', 'scissors'])

  if player == computer:
    print("Draw")
  elif player == "rock" and computer == "scissors":
    print("Player wins!")
  elif player == "scissors" and computer == "paper":
    print("Player wins!")
  elif player == "paper" and computer == "rock":
    print("Player wins!")
  else:
    print ("Computer wins")

  play = input("Do you want to continue (y/n)? ")
```
