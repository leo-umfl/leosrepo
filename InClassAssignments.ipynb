1st ICA

```python
sales = float(input("Sales Amount: "))
tax = float(input("Tax Rate (%): "))
cogs = sales*.4
grossprofit = sales-cogs
taxes = grossprofit*(tax/100)
netincome = grossprofit-taxes

print(f"Sales: {sales:.2f}")
print(f"COGS: {cogs:.2f}")
print(f"Gross Profit: {grossprofit:.2f}")
print(f"Taxes: {taxes:.2f}")
print(f"Net Income: {netincome:.2f}")
```

2nd ICA

```python
number1 = float(input("First Number: "))
operator = (input("Operator: "))
number2 = float(input("Second Number: "))

if operator == "+":
  res = number1+number2
elif operator == "-":
  res = number1-number2
elif operator == "*":
  res = number1*number2
elif operator == "/":
  res = number1/number2
elif operator == "//":
  res = number1//number2
elif operator == "**":
  res = number1**number2
elif operator == "%":
  res= number1%number2
else:
  print("invalid operation")

print(f"Result: {res}")
```
